```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('auto-mpg.csv')
df.info()
```

```python
df['horsepower'].unique()
```

```python
df = df.loc[df['horsepower']!='?']
```

```python
df.info()
```

```python
df['horsepower']=df['horsepower'].astype('int')
```

```python
df.info()
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df['acceleration'])
```

```python
df.describe()
```

```python
q1=df['acceleration'].quantile(0.25)
q3=df['acceleration'].quantile(0.75)
iqr=q3-q1
upper_range = q3 + 1.5*iqr
lower_range = q1 - 1.5*iqr
df = df.loc[(df['acceleration']>lower_range)&(df['acceleration']<upper_range)]
```

```python
df.info()
```

```python
w1=df['horsepower'].quantile(0.25)
w3=df['horsepower'].quantile(0.75)
iqr=w3-w1
upper_range = w3 + 1.5*iqr
lower_range = w1 - 1.5*iqr
df=df.loc[(df['horsepower']>lower_range)&(df['horsepower']<upper_range)]
```

```python
df.info()
```

### Scatter Matrix

```python
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(df,figsize=(15,15))
plt.show()
```
