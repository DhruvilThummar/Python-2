### Merge Function 


```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_id':[10,20,30,40]
})

df2 = pd.DataFrame({
    'dept_id':[10,20,30],
    'dept_name':['HR','IT','Finance']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on='dept_id',how='inner')
```

```python
pd.merge(df1,df2,on='dept_id',how='left')
```

```python
pd.merge(df1,df2,on='dept_id',how='right')
```

```python
pd.merge(df1,df2,on='dept_id',how='outer')
```

```python
pd.merge(df1,df2,how='cross')
```

```python
df1=pd.DataFrame({
    'emp_id':[1,2],
    'name':['A','B'],
    'dept_id':[10,20]
})

df2=pd.DataFrame({
    'd_id':[10,20],
    'dept_name':['HR','IT']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='d_id')
```

```python
df1=pd.DataFrame({
    'emp_id':[1,2,3],
    'name':['A','B','C'],
    'dept_id':[10,20,30]
})

df2=pd.DataFrame({
    'd_id':[10,20],
    'dept_name':['HR','IT']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=['_emp','_dept'])
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer',sort=True)
```

### Groupby (Most Useful Function in Pandas)

```python
data = {
    'Department':['IT','IT','HR','HR','FINANCE','IT'],
    'Employee':['A','B','C','D','E','F'],
    'Salary':[50000,60000,45000,48000,70000,55000],
    'Experiance':[2,3,1,2,5,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby('Department')
```

```python
df.groupby('Department').mean(numeric_only=True)
```

```python
df.drop('Employee',axis=1).groupby('Department').agg(['max','mean','sum'])
```

```python
df.groupby(['Department','Employee']).sum()
```

```python
data = {
    'Department':['IT','IT','HR','HR','FINANCE','IT'],
    'Employee':['A','B','C','D','E','A'],
    'Salary':[50000,60000,45000,48000,70000,55000],
    'Experiance':[2,3,1,2,5,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Department','Employee']).sum()
```

```python
df.groupby('Department',as_index=False,sort=False).sum(numeric_only=True)
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department').agg({'Salary':['max','mean','median'],
                             'Experiance':'mean'})
```

```python
df.groupby('Department').size()
```

```python
df.groupby('Department').count()
```

```python
df.groupby('Department').describe()
```

```python
df.groupby('Department').nth(0)
```

```python
df.groupby('Department').nth(1)
```

```python
def fun(x):
    x['Highest_Earner'] = x['Salary']>x['Salary'].mean()
    return x
df_new=df.groupby('Department').apply(fun)
df_new
```

```python
group=df.groupby('Department')
group.get_group('IT')
```
