```python
data = [10,25,30,35]
ser = pd.Series(data,index=['A','B','C','D'])
print(ser)
```

```python
data = [10,25.0,30,35]
ser = pd.Series(data,index=['A','B','C','D'],name='Series')
print(ser)
```

```python
data = ['Python','FSD','COA','DM']
ser = pd.Series(data,index=['A','B','C','D'],name='Series')
print(ser)
```
