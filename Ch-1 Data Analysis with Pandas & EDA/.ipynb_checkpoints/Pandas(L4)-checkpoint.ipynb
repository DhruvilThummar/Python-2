```python
data = {
    'Name':['A','B','C'],
    'Marks':[30,40,50],
    'Percent':[60,70,80]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.set_index('Name',inplace=True)
df
```

```python
df['Result']=df['Marks'].apply(lambda x:'Pass' if x>=35 else 'Fail')
```

```python
df
```

```python
df['Marks_Extra'] = df['Marks'].apply(lambda x:x+4 if x>=35 else x)
df
```

```python
def grade(x):
    if x['Marks']>60:
        return 'A'
    elif x['Marks']>=40:
        return 'B'
    else:
        return 'F'
df['Grade']=df.apply(grade,axis=1)
df
```

```python
def fun(x,val):
    return x+val
df['Percent']=df['Percent'].apply(fun,args=(5,))
df
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1,inplace=True)
```

```python
df
```

```python
df.sort_values(by='Grade')
```

```python
df.sort_values(by='Marks_Extra',ascending=True)
```

### Concat Function

```python
df1=pd.DataFrame({'A':[1,2],'B':[5,6]})
df2=pd.DataFrame({'A':[3,4],'B':[7,8]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[5,6]})
df2=pd.DataFrame({'A':[3,4],'D':[7,8]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner')
```

```python
df1=pd.DataFrame({'A':[1,2],'E':[5,6]})
df2=pd.DataFrame({'F':[3,4],'D':[7,8]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],sort=True,ignore_index=False)
```

```python
pd.concat([df1,df2],join='inner')
```

```python
pd.concat([df1,df2],join='inner').shape
```
