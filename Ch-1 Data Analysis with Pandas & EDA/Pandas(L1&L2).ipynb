```python
d=pd.read_csv('Book1.csv')
#print(type(d))
ser = d['Marks']
print(ser)
```

```python
print(ser.max())
print(ser.min())
print(ser.size)
print(ser.shape)
print(ser.mean()) # mean is the average of the values in the Series. It is calculated by summing all the values and dividing by the number of values.
print(ser.median()) # median is the middle value when the data is sorted. If there are an even number of values, it is the average of the two middle values.
print(ser.mode()[0]) # mode is the value that appears most frequently in the Series. If there are multiple modes, it returns a Series of all modes. Here we are taking the first mode using [0].
```

```python
data=[
    [1,2,3,5,3],
    [3,3,5,2,1],
    [1,2,3,5,3],
    [1,2,3,5,3],
    [4,5,6,3,6],
    [4,6,3,1,5]
]
df=pd.DataFrame(data)
df
```

```python
df1=df.drop_duplicates()
df1
```

```python
df1=df.drop_duplicates(subset=[0,1,2]) # subset parameter is used to specify the columns based on which duplicates should be identified. Here we are checking for duplicates based on the first three columns (0, 1, and 2).
df1.reset_index(drop=True) # reset_index is used to reset the index of the DataFrame after dropping duplicates. drop=True is used to avoid adding the old index as a new column in the resulting DataFrame.
```

```python
df.duplicated(keep='last') # duplicated() method is used to identify duplicate rows in the DataFrame. keep='last' parameter marks all duplicates as True except for the last occurrence, which is marked as False. This means that if there are multiple duplicate rows, only the last one will be considered unique (not a duplicate), and all previous occurrences will be marked as duplicates (True).`
```
