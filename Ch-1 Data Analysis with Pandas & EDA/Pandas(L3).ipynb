```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('auto-mpg.csv')
df.info()
```

```python
df['horsepower'].unique()
```

```python
df = df.loc[df['horsepower']!='?'] # This line filters the DataFrame to include only rows where the 'horsepower' column does not contain the '?' value. The loc[] method is used to select rows based on a condition, and in this case, we are checking for rows where 'horsepower' is not equal to '?'. This is important because '?' is likely a placeholder for missing or invalid data, and we want to exclude those rows from our analysis.
df
```

```python
df.info()
```

```python
df['horsepower']=df['horsepower'].astype('int') # This line converts the 'horsepower' column from its current data type (which is likely object or string due to the presence of '?') to an integer data type. The astype() method is used to change the data type of a column, and in this case, we are converting 'horsepower' to integers so that we can perform numerical analysis on it.
```

```python
df.info()
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df['acceleration'])
df['acceleration']
```

```python
df.describe()
```

```python
q1=df['acceleration'].quantile(0.25) # This line calculates the first quartile (Q1) of the 'acceleration' column in the DataFrame. The quantile() method is used to compute the specified quantile value, and in this case, we are calculating the 25th percentile (0.25) which represents the value below which 25% of the data falls. Q1 is a measure of central tendency that helps us understand the distribution of the 'acceleration' data.
q3=df['acceleration'].quantile(0.75)  # This line calculates the third quartile (Q3) of the 'acceleration' column in the DataFrame. Similar to Q1, the quantile() method is used to compute the specified quantile value, and in this case, we are calculating the 75th percentile (0.75) which represents the value below which 75% of the data falls. Q3 is another measure of central tendency that helps us understand the distribution of the 'acceleration' data, particularly the upper end of the distribution.
iqr=q3-q1
upper_range = q3 + 1.5*iqr
lower_range = q1 - 1.5*iqr
df = df.loc[(df['acceleration']>lower_range)&(df['acceleration']<upper_range)]
iqr,upper_range,lower_range
```

```python
df.info()
```

```python
w1=df['horsepower'].quantile(0.25)
w3=df['horsepower'].quantile(0.75)
iqr=w3-w1
upper_range = w3 + 1.5*iqr
lower_range = w1 - 1.5*iqr
df=df.loc[(df['horsepower']>lower_range)&(df['horsepower']<upper_range)]
```

```python
df.info()
```

### Scatter Matrix

```python
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(df,figsize=(15,15),alpha=0.5)
plt.show()
```

```python
pd.plotting.parallel_coordinates(df,class_column='cylinders',cols=['mpg','acceleration','origin'],color=['red','blue','green','yellow','orange'])
plt.show()
```
