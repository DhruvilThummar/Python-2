### Merge Function 


```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_id':[10,20,30,40]
})

df2 = pd.DataFrame({
    'dept_id':[10,20,30],
    'dept_name':['HR','IT','Finance']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on='dept_id',how='inner') # This line merges the two DataFrames, df1 and df2, based on the common column 'dept_id'. The on='dept_id' parameter specifies that the merge should be performed using the 'dept_id' column as the key. The how='inner' parameter indicates that only the rows with matching 'dept_id' values in both DataFrames will be included in the resulting merged DataFrame. If there are any 'dept_id' values in df1 that do not have a corresponding match in df2, or vice versa, those rows will be excluded from the final result.
```

```python
pd.merge(df1,df2,on='dept_id',how='left')  # This line merges the two DataFrames, df1 and df2, based on the common column 'dept_id'. The on='dept_id' parameter specifies that the merge should be performed using the 'dept_id' column as the key. The how='left' parameter indicates that all rows from df1 will be included in the resulting merged DataFrame, and matching rows from df2 will be included where available. If there are any 'dept_id' values in df1 that do not have a corresponding match in df2, those rows will still be included in the final result, but the columns from df2 will contain NaN values for those unmatched rows.
```

```python
pd.merge(df1,df2,on='dept_id',how='right')   # This line merges the two DataFrames, df1 and df2, based on the common column 'dept_id'. The on='dept_id' parameter specifies that the merge should be performed using the 'dept_id' column as the key. The how='right' parameter indicates that all rows from df2 will be included in the resulting merged DataFrame, and matching rows from df1 will be included where available. If there are any 'dept_id' values in df2 that do not have a corresponding match in df1, those rows will still be included in the final result, but the columns from df1 will contain NaN values for those unmatched rows.
```

```python
pd.merge(df1,df2,on='dept_id',how='outer') # This line merges the two DataFrames, df1 and df2, based on the common column 'dept_id'. The on='dept_id' parameter specifies that the merge should be performed using the 'dept_id' column as the key. The how='outer' parameter indicates that all rows from both DataFrames will be included in the resulting merged DataFrame. If there are any 'dept_id' values in df1 that do not have a corresponding match in df2, those rows will still be included in the final result with NaN values for the columns from df2. Similarly, if there are any 'dept_id' values in df2 that do not have a corresponding match in df1, those rows will also be included with NaN values for the columns from df1.
```

```python
pd.merge(df1,df2,how='cross')  # This line merges the two DataFrames, df1 and df2, using a cross join. A cross join produces the Cartesian product of the two DataFrames, meaning each row from df1 will be combined with every row from df2. The resulting DataFrame will have a row for each possible combination of rows from the two input DataFrames. The on parameter is not needed for a cross join, as it does not require a common key between the DataFrames. The resulting DataFrame will contain all columns from both df1 and df2, and the number of rows will be equal to the product of the number of rows in df1 and df2.
```

```python
df1=pd.DataFrame({
    'emp_id':[1,2],
    'name':['A','B'],
    'dept_id':[10,20]
})

df2=pd.DataFrame({
    'd_id':[10,20],
    'dept_name':['HR','IT']
})
print(df1)
print(df2)
df1,df2
```

```python
pd.merge(df1,df2,left_on='dept_id',right_on='d_id') # This line merges the two DataFrames, df1 and df2, based on the specified columns. The left_on='dept_id' parameter indicates that the merge should use the 'dept_id' column from df1 as the key, while the right_on='d_id' parameter indicates that the merge should use the 'd_id' column from df2 as the key. The resulting DataFrame will contain all columns from both df1 and df2, and rows will be matched based on the values in these specified columns. If there are any values in 'dept_id' from df1 that do not have a corresponding match in 'd_id' from df2, or vice versa, those rows will be excluded from the final result by default (inner join).
```

```python
df1=pd.DataFrame({
    'emp_id':[1,2,3],
    'name':['A','B','C'],
    'dept_id':[10,20,30]
})

df2=pd.DataFrame({
    'd_id':[10,20],
    'dept_name':['HR','IT']
})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True) # This line merges the two DataFrames, df1 and df2, based on their index. The left_index=True parameter indicates that the merge should use the index of df1 as the key, while the right_index=True parameter indicates that the merge should use the index of df2 as the key. The resulting DataFrame will contain all columns from both df1 and df2, and rows will be matched based on their index values. If there are any index values in df1 that do not have a corresponding match in df2, or vice versa, those rows will be excluded from the final result by default (inner join).
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer') 
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=['_emp','_dept']) # This line merges the two DataFrames, df1 and df2, based on their index, similar to the previous merge. The left_index=True and right_index=True parameters indicate that the merge should use the index of both DataFrames as the key. The suffixes=['_emp','_dept'] parameter specifies that if there are any overlapping column names between df1 and df2 (other than the index), they will be renamed in the resulting merged DataFrame by appending '_emp' to columns from df1 and '_dept' to columns from df2. This helps to avoid confusion and allows you to distinguish between columns that originated from each DataFrame in the merged result.
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer',sort=True) 
```

### Groupby (Most Useful Function in Pandas)

```python
data = {
    'Department':['IT','IT','HR','HR','FINANCE','IT'],
    'Employee':['A','B','C','D','E','F'],
    'Salary':[50000,60000,45000,48000,70000,55000],
    'Experiance':[2,3,1,2,5,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby('Department')
```

```python
df.groupby('Department').mean(numeric_only=True)
```

```python
df.drop('Employee',axis=1).groupby('Department').agg(['max','mean','sum'])  # This line of code performs a groupby operation on the DataFrame df based on the 'Department' column. It first drops the 'Employee' column from the DataFrame using df.drop('Employee', axis=1), which removes the specified column. Then, it groups the remaining data by the 'Department' column using groupby('Department'). Finally, it applies the agg function to calculate the maximum (max), mean (mean), and sum (sum) for each numeric column in the grouped DataFrame. The resulting output will show these aggregated statistics for each department, excluding the 'Employee' column from the calculations.
```

```python
df.groupby(['Department','Employee']).sum()
```

```python
data = {
    'Department':['IT','IT','HR','HR','FINANCE','IT'],
    'Employee':['A','B','C','D','E','A'],
    'Salary':[50000,60000,45000,48000,70000,55000],
    'Experiance':[2,3,1,2,5,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Department','Employee']).sum()
```

```python
df.groupby('Department',as_index=False,sort=False).sum(numeric_only=True) # This line of code performs a groupby operation on the DataFrame df based on the 'Department' column. The as_index=False parameter indicates that the 'Department' column should not be set as the index in the resulting grouped DataFrame, allowing it to remain as a regular column. The sort=False parameter specifies that the groups should not be sorted by the 'Department' values in the output. Finally, the sum(numeric_only=True) function is applied to calculate the sum of all numeric columns for each department, while ignoring any non-numeric columns. The resulting output will show the total salary and experience for each department without sorting or setting 'Department' as the index.
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department').agg({'Salary':['max','mean','median'],
                             'Experiance':'mean'}) # This line of code performs a groupby operation on the DataFrame df based on the 'Department' column. It then applies the agg function to calculate specific aggregated statistics for the 'Salary' and 'Experiance' columns. For the 'Salary' column, it calculates the maximum (max), mean (mean), and median (median) values for each department. For the 'Experiance' column, it calculates the mean value for each department. The resulting output will show these aggregated statistics for each department, allowing you to analyze the salary distribution and average experience within each department.
```

```python
df.groupby('Department').size()
```

```python
df.groupby('Department').count()
```

```python
df.groupby('Department').describe()
```

```python
df.groupby('Department').nth(0) # This line of code performs a groupby operation on the DataFrame df based on the 'Department' column and then selects the first row (nth(0)) from each group. The nth() function is used to retrieve a specific row from each group, where the argument specifies the position of the row to be selected. In this case, nth(0) selects the first row from each department group. The resulting output will show one representative row for each department, which can be useful for quickly inspecting the data or for further analysis.
```

```python
df.groupby('Department').nth(1)
# The nth(1) function selects the second row from each department group in the DataFrame df. Similar to nth(0), which selects the first row, nth(1) retrieves the row at index position 1 (the second row) for each group defined by the 'Department' column. The resulting output will show one representative row for each department, specifically the second entry for each department group. If a department has fewer than two entries, it will return NaN for that department in the output.
```

```python
def fun(x):
    x['Highest_Earner'] = x['Salary']>x['Salary'].mean()
    return x
df_new=df.groupby('Department').apply(fun)
df_new
```
