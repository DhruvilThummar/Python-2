```
 The following dictionary shows how five people follow each other on Instagram:
 instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
 E.g., the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be 
added from person1 to person3.
 
 Using networkx library, create a directed graph

````python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

instagram = {'person1': [0,1,1,0,1],
              'person2': [0,0,1,0,1], 
              'person3': [1,1,0,1,1],
              'person4': [1,1,1,0,0],
              'person5': [1,1,0,0,0]}
print(list(instagram.values()))
val = np.array(list(instagram.values()))
G = nx.DiGraph(val)
nx.draw_networkx(G,labels={0:'Person1',1:'Person2',2:'Person3',3:'Person4',4:'Person5'})
plt.show()
````

``` Using Networkx Library, Write a Code to create below given Directed Graph.```

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([(0, 1),(0, 3),(1, 1),(1, 2),(2, 0),(2, 4),(5, 1),(5, 2)])
nx.draw_networkx(G,node_color='pink')
plt.show()
````
