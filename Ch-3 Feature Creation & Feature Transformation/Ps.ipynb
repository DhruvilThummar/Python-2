## Columns and Their Descriptions 
- ID: Unique match identifier. 
- City: City where the match was played (may contain missing values). 
- Date: Date of the match. 
- Season: IPL season year (e.g., 2008, 2022). 
- MatchNumber: Match number in the respective season. 
- Team1: Name of the first team in the match. 
- Team2: Name of the second team in the match. 
- Venue: Stadium or ground where the match was held. 
- TossWinner: Team that won the toss. 
- TossDecision: Toss decision taken (bat or bowl). 
- SuperOver: Indicates if a Super Over was played (Yes/No). 
- WinningTeam: Team that won the match. 
- WonBy: Indicates if the match was won by runs or wickets. 
- Margin: Margin of victory (runs or wickets). 
- method: Method used to determine results in interrupted matches (e.g., Duckworth-Lewis, rarely present). 
- Player_of_Match: Player who was awarded "Player of the Match". 
- Team1Players: List of players in Team 1. 
- Team2Players: List of players in Team 2. 
- Umpire1: Name of the first on-field umpire. 
- Umpire2: Name of the second on-field umpire. 
 

````python
import pandas as pd
import numpy as np

df = pd.read_csv('ipl-matches.csv')
df
````

### Standardization of Team Names (Franchise Renaming) 
- In  the  given  dataset,  some  IPL  teams  are  listed  using  their  old  franchise  names. 
- You are required to standardize the team names by replacing them with their updated official names. 
- ``` Old Name → New Name Mapping:  ```
    - Delhi Daredevils → Delhi Capitals  
    - Kings XI Punjab → Punjab Kings  
    - Rising Pune Supergiants → Rising Pune Supergiant

````python
df.replace(['Delhi Daredevils','Kings XI Punjab','Rising Pune Supergiants'],
          ['Delhi Capitals','Punjab Kings','Rising Pune Supergiant'],
          inplace=True)
df
````

### Identify the Top Winning Team per Season 
- For every season: 
    - Identify the team with the highest number of wins  
    -  Display:  
        - Season  
        - Winning Team (with maximum wins)  
        - Number of Wins 
    - Output format (DataFrame columns): 

``` Season | WinningTeam | number_of_wins ```

````python
season_wins = (
    df.dropna(subset=['WinningTeam'])
      .groupby(['Season', 'WinningTeam'])
      .size()
      .reset_index(name='number_of_wins')
      .sort_values(['Season', 'number_of_wins'], ascending=[True, False])
      .drop_duplicates(subset='Season')
      .reset_index(drop=True)
)

season_wins
````

### Player of the Match Diversity per Winning Team 
- For each winning team: 
    - Calculate the number of unique players who received the Player of the Match award  
    - Display the diversity count for each team  
    - Arrange the results in descending order of diversity 
    - Output format (DataFrame columns): 
 
``` WinningTeam | number_of_unique_player ```

````python
ab=df.groupby('WinningTeam')['Player_of_Match'].nunique().sort_values(ascending=False).reset_index(name='number_of_unique_players')
ab
````

### Display Teams by Match Wins Based on Toss Decision (Descending Order )
- For each toss decision: 
    - Calculate the number of matches won by each team. 
    - Display the results in descending order of wins. 
    - Output format (DataFrame columns): 
 
``` TossDecision | WinningTeam | wins ```

````python
ab = (
    df.groupby(['TossDecision', 'WinningTeam'])
      .size()
      .reset_index(name='wins')
      .sort_values(['TossDecision', 'wins'], ascending=[True, False])
      .reset_index(drop=True)
)
ab
````

### Analysis of Top Five Venues with Highest Match Frequency and Their Dominant Winning Teams 
(Descending Order) 
-  Identify the top 5 venues where the highest number of matches have been played 
-  For each of these venues: 
    - Calculate which team has won the most matches at that venue 
    - Display the dominant team along with the number of wins 
        -  Sort the results in descending order of winning matches 
        - Output format (DataFrame columns): 
        
``` Venue | WinningTeam | no_of_winning_match ```

````python
top_venues = df['Venue'].value_counts().head(5).index

venue_dominant_team = (
    df[df['Venue'].isin(top_venues)]
    .dropna(subset=['WinningTeam'])
    .groupby(['Venue', 'WinningTeam'])
    .size()
    .reset_index(name='no_of_winning_match')
    .sort_values(['Venue', 'no_of_winning_match'], ascending=[True, False])
    .drop_duplicates(subset='Venue')
    .sort_values('no_of_winning_match', ascending=False)
    .reset_index(drop=True)
)

venue_dominant_team
````

### All Season-wise IPL Points Table Generation 
- For each season output in data frame: 
    - Calculate the total number of matches played by each team  
    - Calculate the total number of matches won by each team  
    - Identify matches with no result  
    - Compute total points using:  
        - Win = 2 points  
        - No Result = 1 point  
    - sorting teams based on points in descending order 
- Output format (example): 
``` 
Season: 2022                                                               
Team Name                  Matches Played  Matches Won  No Result  Points  
Gujarat Titans                         16           12          0      24  

```  

````python
# Season-wise points table

matches_played = (
    df.melt(id_vars=['Season'], value_vars=['Team1', 'Team2'], value_name='Team')
      .groupby(['Season', 'Team'])
      .size()
      .reset_index(name='Matches Played')
)

matches_won = (
    df.dropna(subset=['WinningTeam'])
      .groupby(['Season', 'WinningTeam'])
      .size()
      .reset_index(name='Matches Won')
      .rename(columns={'WinningTeam': 'Team'})
)

no_result = (
    df[df['WinningTeam'].isna()]
      .melt(id_vars=['Season'], value_vars=['Team1', 'Team2'], value_name='Team')
      .groupby(['Season', 'Team'])
      .size()
      .reset_index(name='No Result')
)

points_table = (
    matches_played
    .merge(matches_won, on=['Season', 'Team'], how='left')
    .merge(no_result, on=['Season', 'Team'], how='left')
    .fillna(0)
)

points_table[['Matches Won', 'No Result']] = points_table[['Matches Won', 'No Result']].astype(int)
points_table['Points'] = points_table['Matches Won'] * 2 + points_table['No Result']

points_table = (
    points_table
    .rename(columns={'Team': 'Team Name'})
    .sort_values(['Season', 'Points', 'Matches Won'], ascending=[True, False, False])
    .reset_index(drop=True)
)

# Optional: display season-wise like the requested format
for season, season_df in points_table.groupby('Season', sort=True):
    print(f"Season: {season}")
    display(season_df[['Team Name', 'Matches Played', 'Matches Won', 'No Result', 'Points']].reset_index(drop=True))
````
